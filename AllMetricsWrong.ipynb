```python
import pandas as pd
import numpy as np

# Creating the dataset
data = [0.05, 0.02, 0.08, 0.04, 0.06] * 19 # 95 background samples
data.extend([0.1, 0.09, 0.07])             # 3 more background
data.append(500.0)                         # The "Nugget" (Outlier)
data.append(-99.0)                         # The "Error" (Sensor failure)

df = pd.DataFrame(data, columns=['Au_Grade'])

# Calculating 4 metrics
mean_val = df['Au_Grade'].mean()
median_val = df['Au_Grade'].median()
range_val = df['Au_Grade'].max() - df['Au_Grade'].min()
success_rate = (df['Au_Grade'] > 1.0).sum() / len(df) * 100

print(f"--- THE SAME DATASET ---")
print(f"A. Mean Grade:    {mean_val:>8.2f} g/t  -> 'Looks profitable!'")
print(f"B. Median Grade:  {median_val:>8.2f} g/t  -> 'Looks like a waste of time'")
print(f"C. Range:         {range_val:>8.2f} g/t  -> 'Looks like total chaos'")
print(f"D. Success Rate:  {success_rate:>8.2f} %    -> 'Looks like a failure'")
```
